<h4>Objective :</h4>
    We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

```python
# Import required libraries
import cv2
```

```python
# Set the working directory
import io
%cd "E:\LetsGrowMore\Tasks\Task 4 - Image to Pencil Sketch with Python"
```

```python
# Now we will read the image
lion_image = cv2.imread(r'E:\LetsGrowMore\Tasks\Task 4 - Image to Pencil Sketch with Python\Lion.jpg',1)
print(lion_image.shape)
```

As you can see that the image size is huge so we will resize it so that we can see it properly.

```python
# Resize the image
lion_image = cv2.resize(lion_image, dsize=(0,0), fx = 0.2, fy = 0.2)
print(lion_image.shape)
cv2.imshow("Lion", lion_image)
cv2.waitKey(0)
```

```python
# Now we will convert this image to grayscale
lion_gray = cv2.cvtColor(lion_image, cv2.COLOR_RGB2GRAY)
cv2.imshow("Lion_Gray", lion_gray)
cv2.waitKey(0)
```

```python
# Now we will invert the grayscale image
lion_gray_inverted = 255 - lion_gray
cv2.imshow("lion_gray_inverted", lion_gray_inverted)
cv2.waitKey(0)
```

```python
# Now we blur the image by using the Gaussianblur Function in cv2
lion_blurred = cv2.GaussianBlur(lion_gray_inverted, (21,21), 0)
cv2.imshow("lion_blurred", lion_blurred)
cv2.waitKey(0)
```

```python
# Now we will invert the blurred image and then convert it to a pencil sketch using divide function in cv2
lion_blurred_inverted = 255 - lion_blurred
cv2.imshow("lion_blurred_inverted", lion_blurred_inverted)
cv2.waitKey(0)
```

```python
# convert blurred image to a pencil sketch using divide function in cv2
lion_sketch = cv2.divide(lion_gray, lion_blurred_inverted, scale=256.0)
cv2.imshow("lion_sketch", lion_sketch)
cv2.waitKey(0)
```

```python
# Now we will look at both original image and the pencil sketch
cv2.imshow("Original", lion_image)
cv2.imshow("Pencil Sketch", lion_sketch)
cv2.waitKey(0)
```

```python
# Blurred image
plt.figure(figsize=(8,5))
plt.imshow(cv2.cvtColor(lion_blurred,cv2.COLOR_RGB2BGR))
plt.axis('off')
plt.title('Blurred', size = 15)
plt.show()
```

```python
# inverted blurred image
plt.figure(figsize=(8,5))
plt.imshow(cv2.cvtColor(lion_blurred_inverted,cv2.COLOR_RGB2BGR))
plt.axis('off')
plt.title('Blurred Inverted', size = 15)
plt.show()
```

```python
# pencil sketch
plt.figure(figsize=(8,5))
plt.imshow(cv2.cvtColor(lion_sketch,cv2.COLOR_RGB2BGR))
plt.axis('off')
plt.title('Pencil Sketch', size = 15)
plt.show()
```
